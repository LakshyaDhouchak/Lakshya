```python
from langgraph.graph import StateGraph,START ,END
from typing import TypedDict
from langchain_ollama import ChatOllama
from dotenv import load_dotenv
from langgraph.checkpoint.memory import InMemorySaver
```

```python
load_dotenv()
model =  ChatOllama(model = 'minimax-m3:cloud', temperature= 0.5)
```

```python
# define the state Class
class JockState(TypedDict):
    # define the properties
    topic : str
    joke : str
    explanation : str

# define the methord
def generate_jock(State: JockState):
    # define the joke
    prompt = f'generate a joke on the topic {State["topic"]}'
    response = model.invoke(prompt).content
    return {'joke': response}

def generate_explanation(State: JockState):
    # prompt
    prompt = f'Write an explanation for the jock - {State["joke"]}'
    response = model.invoke(prompt).content
    return {'explanation': response}
```

```python
graph = StateGraph(JockState)

# add node
graph.add_node('generate_jock', generate_jock)
graph.add_node('generate_explanation', generate_explanation)

# add edge
graph.add_edge(START,'generate_jock')
graph.add_edge('generate_jock','generate_explanation')
graph.add_edge('generate_explanation', END)

checkPointer = InMemorySaver()
workflow = graph.compile(checkpointer= checkPointer)
workflow
```

```python
config1 = {'configurable':{'thread_id':'1'}}
workflow.invoke({'topic':'pizza'}, config= config1)
```

```python
workflow.get_state(config1)
```
