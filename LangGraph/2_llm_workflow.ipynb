```python
from langchain_ollama import ChatOllama
from langgraph.graph import StateGraph, START, END
from dotenv import load_dotenv
from typing import TypedDict
```

```python
load_dotenv()
model =  ChatOllama(model = 'minimax-m3:cloud', temperature= 0.5)
```

```python
# create a state
class LLMState(TypedDict):
    question: str
    answer: str

def llm_question(state: LLMState)-> LLMState :
    question = state['question'] 

    prompts = f'Answer the following Question{question}' 
    answer = model.invoke(prompts).content

    state['answer'] = answer
    return state
```

```python
# create a graph
graph = StateGraph(LLMState)

# add node
graph.add_node('llm_question',llm_question)

# add edge
graph.add_edge(START,'llm_question')
graph.add_edge('llm_question',END)

workflow = graph.compile()

intial_state = {'question': 'How far is moon from the earth?'}
final_state = workflow.invoke(intial_state)
print(final_state['answer'])


```
