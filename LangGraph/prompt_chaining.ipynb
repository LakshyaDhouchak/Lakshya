```python
from langchain_ollama import ChatOllama
from dotenv import load_dotenv
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
load_dotenv()
model =  ChatOllama(model = 'minimax-m3:cloud', temperature= 0.5)
```

```python
class BlogState(TypedDict):
    title: str
    outline : str
    content : str

# Define the methord
def createOutline(state: BlogState)-> BlogState:
    title = state['title']

    prompt = f'Generate a detailed outline for a blog on the topic - {title}'
    outline = model.invoke(prompt).content
    state['outline'] = outline

    return state

def createBlog(state: BlogState) -> BlogState:
    title = state['title']
    outline = state['outline']

    prompt = f'Write a detailed blog on the title - {title} using the follwing outline \n {outline}'
    content = model.invoke(prompt).content

    state['content'] =  content
    return state

```

```python
graph =  StateGraph(BlogState)

# add node
graph.add_node('create_outline', createOutline)
graph.add_node('create_blog',createBlog)

# add edge
graph.add_edge(START,'create_outline')
graph.add_edge('create_outline', 'create_blog')
graph.add_edge('create_blog', END)

workflow = graph.compile()
intial_state = {'title': 'Rise of AI in India'}
final_state = workflow.invoke(intial_state)
print(final_state)
```
