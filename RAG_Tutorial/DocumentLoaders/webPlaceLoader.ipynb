```python
from langchain_community.document_loaders import WebBaseLoader
from langchain_ollama import ChatOllama
from dotenv import load_dotenv
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
load_dotenv()
model =  ChatOllama(model = "minimax-m3:cloud", temperature= 0.5)
parser =  StrOutputParser()
prompt =  ChatPromptTemplate([
    ('human','Answer the following question \n {question} from the following text- \n {text}')
])

url= "https://medium.com/@davidmles/comparing-ssgs-static-site-generators-for-my-blog-a6724a719150"

loader = WebBaseLoader(url)
docs = loader.load()
```

```python
chain =  prompt | model | parser
result =  chain.invoke({'question':'What is the prodcut that we are talking about?', 'text':docs[0].page_content})
print(result)
```
