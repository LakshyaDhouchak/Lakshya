```python
from langchain_ollama import OllamaEmbeddings
from dotenv import load_dotenv

load_dotenv()

EmbeddingModels = OllamaEmbeddings(model="twwch/m3e-base:latest", dimensions= 36)
VectorResult = EmbeddingModels.embed_query("Delhi is the capital of India")
print(VectorResult)
```
