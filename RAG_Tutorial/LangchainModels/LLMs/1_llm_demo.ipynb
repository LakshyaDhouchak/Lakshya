NOTE: the LLMs take the string input and give the String Output as result

In this code dotenv byDefault Search for OPENAI_API_KEY so we don't have to specify

```python
from langchain_openai import OpenAI
# This library is used to load API Key from env to individual file
from dotenv import load_dotenv

load_dotenv()

# Create the OpenAI Object 
llm = OpenAI(model ='gpt-3.5-turbo-instruct')
result = llm.invoke("What is the capital of India")

print(result)
```

if i want to give the custom name then how to do that

```python
import os
from langchain_openai import OpenAI
from dotenv import load_dotenv

load_dotenv()
custom_key = os.getenv("Lakshya")
llm = OpenAI(model = 'gpt-3.5-turbo-instruct', api_key= custom_key)
result = llm.invoke("Capital of India ?")
print(result)
```
