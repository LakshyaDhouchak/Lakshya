```python
from langchain_ollama import ChatOllama
from dotenv import load_dotenv
from langchain_core.output_parsers import JsonOutputParser
from langchain_core.prompts import ChatPromptTemplate
```

```python
load_dotenv()
model = ChatOllama(model = "minimax-m3:cloud", temperature= 0.5)
parser = JsonOutputParser()
```

```python
template1 = ChatPromptTemplate([
    ('human','Give the pros and cons on {topic} \n {format_Instruction}')
])

chain = template1 | model | parser
result = chain.invoke({'topic':'black hole','format_Instruction': parser.get_format_instructions()}) 
print(result)
```
