```python
from langchain_ollama import ChatOllama
from dotenv import load_dotenv
from langchain_core.output_parsers import PydanticOutputParser
from pydantic import BaseModel , Field
from langchain_core.prompts import ChatPromptTemplate
```

```python
load_dotenv()
model = ChatOllama(model="minimax-m3:cloud",temperature= 0.5)

class Person(BaseModel):
    name: str = Field(description="Name of the person")
    age: int =Field(gt=18, description=" Age of the person")
    city: str = Field(description="Name of the city the person belong to")

parser = PydanticOutputParser(pydantic_object=Person)    
```

```python
prompt = ChatPromptTemplate([
    ('human','Generate the name, age, city of the fictional {place} person \n {format_Instruction}')
])

chain = prompt | model | parser
result = chain.invoke({'place':'England','format_Instruction': parser.get_format_instructions})
print(result)
```
