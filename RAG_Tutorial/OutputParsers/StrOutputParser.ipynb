```python
from langchain_ollama import ChatOllama
from dotenv import load_dotenv
from langchain_core.prompts import ChatPromptTemplate
```

```python
load_dotenv()
model = ChatOllama(model="minimax-m3:cloud",temperature=0.5)
```

```python
# first prompt -> detailed report
template1 = ChatPromptTemplate([
    ('human',"write a detailed report on {topic}")
])

template2 = ChatPromptTemplate([
    ('human',"write a 5 line summary of the following text.\n {text}")
])
# second prompt -> summary
```

```python
chain1 = template1 | model
result = chain1.invoke({'topic':'black hole'})

chain2 = template2 | model
output = chain2.invoke({"text":result.content})
```

```python
print("Report:\n")
print(result.content)
print("\nSUMMARY\n")
print(output.content)
```
