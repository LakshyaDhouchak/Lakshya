```python
from langchain_ollama import ChatOllama
from dotenv import load_dotenv
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
load_dotenv()
model = ChatOllama(model="minimax-m3:cloud",temperature=0.5)
parser = StrOutputParser()
```

```python
# first prompt -> detailed report
template1 = ChatPromptTemplate([
    ('human',"write a detailed report on {topic}")
])

template2 = ChatPromptTemplate([
    ('human',"write a 5 line summary of the following text.\n {text}")
])
# second prompt -> summary
```

```python
chain = template1 | model | parser | template2 | model | parser
result = chain.invoke({'topic':'black hole'})
print(result)
```
