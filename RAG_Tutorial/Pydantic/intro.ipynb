```python
from pydantic import BaseModel
```

```python
class Person(BaseModel):
    name:str
    age:int
    city:str

person = Person(name = "Lakshya", age = 21, city = "Sonipat")   
print(person) 
```

the dataclasses models don't handle the validation but in case of pydantic is does.

```python
from dataclasses import dataclass
@dataclass
class Person():
    name:str
    age:int
    city:str
person = Person(name = "Lakshya", age = 21, city = "Sonipat")   
print(person) 
```

Model with optional filds
if we use optinal parameter then it do auto typecasting

```python
from typing import Optional
class Employee(BaseModel):
    id:int
    name:str
    department:str
    salary:Optional[float] = None
    is_active:Optional[bool] = True
emp1 = Employee(id=1, name="Lakshya", department= "IT")
print(emp1)    

```

```python
from typing import List


class ClassRoom(BaseModel):
    room_number:str
    student:List[str]
    capacity: int

classroom = ClassRoom(room_number="A101", student=["Lakshya","Deevya","Pavi"], capacity=30)
print(classroom)    
```

Model with nested Models

```python
from pydantic import BaseModel

class Address(BaseModel):
    street:str
    city:str
    zip_code:str

class Customer(BaseModel):
    customer_id:int  
    name:str
    address: Address  

customer =Customer(customer_id=1, name="Emma", address={"street": "123 Main St", "city":"Lakshya", "zip_code":"131001"})  
print(customer)
```

Pydantic Fields: customization and constraints

```python
from pydantic import BaseModel, Field
class Item(BaseModel):
    name:str=Field(min_length=2, max_length=50)
    price:float= Field(gt=0,le=1000)
    quantity:int=Field(ge=0)

item = Item(name="Book", price=29.99, quantity=10)  
print(item)  
```

```python
from pydantic import BaseModel
class User(BaseModel):
    username:str=Field(description="unique username for the user")
    age: int =Field(default=18,description="User age, default to 18")
    email: str = Field(default_factory=lambda:"user@example.com",description="default email address")

User1= User(username="alice") 
print(User1)  

User2= User(username="Lakshya", age = 14, email="lakshya10171@gmail.com")
print(User2)
```
