```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import ChatPromptTemplate
from dotenv import load_dotenv
from langchain_core.runnables import RunnableAssign , RunnablePassthrough, RunnableLambda, RunnableSequence
```

```python
load_dotenv()
model =  ChatOllama(model = "minimax-m3:cloud", temperature= 0.5)
prompt1 = ChatPromptTemplate([
    ('human','Generate a jock about {topic}')
])
length_chain = RunnableLambda(lambda x: {"length": len(x["topic"])})
```

```python
chain = RunnableSequence(RunnablePassthrough(),RunnableAssign({'info':length_chain}))
result = chain.invoke({'topic':'AI'})
print(result)
```
