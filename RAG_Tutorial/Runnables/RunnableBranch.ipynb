```python
from langchain_ollama import ChatOllama
from dotenv import load_dotenv
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableBranch, RunnableSequence, RunnableLambda, RunnablePassthrough
```

```python
load_dotenv()
model = ChatOllama(model = "minimax-m3:cloud", temperature= 0.5)
parser =  StrOutputParser()
prompt1 =  ChatPromptTemplate([
    ('human','Write a detailed report on {topic}')
])
prompt2 =  ChatPromptTemplate([
    ('human','Summarize the following text\n{text}')
])
```

```python
chain =  RunnableSequence(prompt1, model, parser)
branch_chain  =  RunnableBranch(
    (lambda x: len(x.split())>500, RunnableSequence(prompt2, model,parser)),
    RunnablePassthrough()
)
final_chain = RunnableSequence(chain,branch_chain)
result = final_chain.invoke({'topic': "AI"})
print(result)
```
