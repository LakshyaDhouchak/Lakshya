```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv
from langchain_core.runnables import RunnableSequence, RunnableLambda, RunnablePassthrough, RunnableParallel
```

```python
load_dotenv()
model = ChatOllama(model = "minimax-m3:cloud", temperature= 0.5)
parser =  StrOutputParser()
prompt = ChatPromptTemplate([
    ('human','Generate a jock on {topic}')
])

length_words = RunnableLambda(lambda x: {'length': len(x['topic'])} )
```

```python
chain = RunnableSequence(prompt,model,parser)
chain_parallel = RunnableParallel({
    'jock': chain,
    'words': length_words
})

result =  chain_parallel.invoke({'topic':'AI'})
print(result)
```

Runnable Lambda

```python
from langchain_core.runnables import RunnableLambda

def word_counter(text):
    return len(text.split())

runnable_word_counter =  RunnableLambda(word_counter)
result = runnable_word_counter.invoke("Hii how are you")
print(result)
```
