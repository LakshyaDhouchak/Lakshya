```python
from langchain_ollama import ChatOllama
from dotenv import load_dotenv
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough, RunnableSequence, RunnableParallel
```

```python
load_dotenv()
model =  ChatOllama(model = 'minimax-m3:cloud', temperature= 0.5)
prompt1 = ChatPromptTemplate([
    ('human','write a jock about the {topic}')
])
prompt2 = ChatPromptTemplate([
    ('human','now explain this {jock}')
])
parser = StrOutputParser()
```

```python
chain = RunnableSequence(prompt1, model, parser)
chain_parallel = RunnableParallel({
    'jock': RunnablePassthrough(),
    'Explain': RunnableSequence(prompt2, model, parser)
})

final_chain =  RunnableSequence(chain, chain_parallel)
result = final_chain.invoke({'topic':'AI'})
print(result)
```
