```python
from langchain_text_splitters import CharacterTextSplitter

text = """
Beneath the sun or floodlight's gleam,Cricket lives like a waking dream.
A field of green, a willowed sound,Where legends rise and tales are found.From dusty lanes where barefoot boys,
Chase every run with shrieks of joy,
To packed arenas roaring loud,
"""

splitter = CharacterTextSplitter(
    separator='',
    chunk_size = 50,
    chunk_overlap = 0

)
result = splitter.split_text(text)
print(result)
```

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import CharacterTextSplitter

loader = PyPDFLoader('dl-curriculum.pdf')
docs = loader.load()

splitter = CharacterTextSplitter(
    separator='',
    chunk_size = 100,
    chunk_overlap = 0
)

result = splitter.split_documents(docs)
print(result)
```
