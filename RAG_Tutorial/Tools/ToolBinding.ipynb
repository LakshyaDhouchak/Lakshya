```python
from langchain_ollama import ChatOllama
from langchain_core.messages import HumanMessage
from langchain_core.tools import tool
import requests
```

```python
# define the tools
@tool
def multiply(a:int , b:int) -> int:
    """ Given two number a and b this tool returns thier product """
    return a*b
print(multiply.invoke({'a':3,'b':2}))
```

```python
model =  ChatOllama(model = "minimax-m3:cloud", temperature= 0.5)
model.invoke('hii')
```

```python
# tools binding
llm_with_tools = model.bind_tools([multiply])
llm_with_tools.invoke("hii How are you")
```

```python
query = HumanMessage("can you multiply 3 by 1000")
message = [query]
result = llm_with_tools.invoke(message)
message.append(result)
```

```python
tool_result = multiply.invoke(result.tool_calls[0])
print(tool_result)
```
